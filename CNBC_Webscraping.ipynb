```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
linklist = []
# read the file line by line
with open('cndn_linklist.txt', 'r') as file:
    lines = [line.strip("'").rstrip(",\n'") for line in file]

# append each line into the linklist list.
for line in lines:
    linklist.append(line)
```

```python
# Check the link content
linklist
```

```python
data = []
```

```python
# Read the headline and time of each link and append the data into the data list.
for link in linklist:
    url = link
    response = requests.get(url)
    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')
    headline_tag = soup.find('h1', class_='ArticleHeader-headline')
    headline = headline_tag.get_text(strip=True) 
    published_time_tag = soup.find('time', {'data-testid': 'published-timestamp'})
    published_datetime = published_time_tag['datetime'] 
    print(published_datetime,headline)
    data.append({'time':  published_datetime, 'review': headline})
    
```

```python
#Convert data into dataframe
gamestop_cnbc = pd.DataFrame(data)
```
